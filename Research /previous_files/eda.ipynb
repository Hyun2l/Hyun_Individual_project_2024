```python
import pandas as pd

# Assuming df is your DataFrame

# Check the dtype of the 'LapTime' column
print(df['LapTime'].dtype)

# If 'LapTime' is not a timedelta, convert it
if df['LapTime'].dtype != 'timedelta64[ns]':
    df['LapTime'] = pd.to_timedelta(df['LapTime'])

# Now you can safely convert to total seconds
df['LapTime'] = df['LapTime'].dt.total_seconds()


```

```python
df["LapTime"].info()

df["LapTime"].plot()


```

```python

numeric_columns = df.select_dtypes(include=[np.number])
medians = numeric_columns.median(numeric_only=True)
df = df.fillna(medians)


df = df.dropna()
# Step 2: Drop Rows with Any Remaining Missing Values
# !! we drop rows that still have any missing values, which might not be necessary if the previous step addresses all missing values, but it's a good practice to ensure the DataFrame is clean.


mean = df[numeric_columns.columns].mean()
std = df[numeric_columns.columns].std()
z_scores = (df[numeric_columns.columns] - mean) / std


df_filtered = df[(z_scores < 3).all(axis=1)]

```

```python
# Check for any entirely NaN columns
nan_columns = df.isna().all(axis=0)
print("Columns with all NaN values:", nan_columns[nan_columns].index.tolist())

# Check for any entirely NaN rows
nan_rows = df.isna().all(axis=1)
print("Number of rows with all NaN values:", nan_rows.sum())
```

```python
# Histograms for numerical data
df.hist(bins=15, figsize=(15, 10), layout=(5, 4))
plt.show()

# Boxplots for weather conditions to identify outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['AirTemp', 'Humidity', 'Pressure', 'TrackTemp']])
plt.show()

```

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Heatmap to visualize correlations
plt.figure(figsize=(20, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

# Scatter plots for specific weather conditions vs. lap time
plt.figure(figsize=(12, 6))
sns.scatterplot(x='TrackTemp', y='LapTime', data=df)
plt.title('Track Temperature vs. Lap Time')
plt.show()
```
